```python
#1.1生成Numpy數組
import numpy as np
```

```python
#1.1.1從已有數據中創建數組
lst1 = [3.14, 2.17, 0, 1, 2]
nd1 = np.array(lst1)
type(nd1)
```

```python
lst2 = [[3.14, 2.17], [0, 1, 2]]
nd2 = np.array(lst2)
type(nd2)
```

```python
#1.1.2利用random模塊產生數組

nd3 = np.random.random([3, 3])
print(nd3)
type(nd3.shape)
```

```python
np.random.seed(123)
nd4 = np.random.randn(2, 3)
print(nd4)
np.random.shuffle(nd4)
print(nd4)
```

```python
#1.1.3創建特定形狀的多維數組
nd5 = np.zeros([3, 3])
nd6 = np.ones([3, 3])
nd7 = np.eye(3)
nd8 = np.diag([1, 2, 3])
print(nd5)
print(nd6)
print(nd7)
print(nd8)
```

```python
#1.1.4利用arrange, linspace函數生成數組
print(np.arange(10))
print(np.arange(0, 10))
print(np.arange(1, 4, 0.5))
print(np.arange(9, -1, -1))
```

```python
np.linspace(0, 1, 10)
```

```python
#1.2獲取元素
np.random.seed(2019)
nd11 = np.random.random([10])
nd11[3]
nd11[3:6]
nd11[1:6:2]
nd11[::-2]

nd12 = np.arange(25).reshape([5, 5])
```

```python
#1.3.1對應元素相乘
A = np.array([[1, 2], [-1, 4]])
B = np.array([[2, 0], [3, 4]])

print(A * B)
np.multiply(A, B)
```

```python
print(A*2.0)
A/2.0
```

```python
#1.3.2點積運算
X1 = np.array([[1, 2], [3, 4]])
X2 = np.array([[5, 6], [7, 8, 9]])
np.dot(X1, X2)
```

```python
#1.4數組變形
#1.reshape
arr = np.arange(10)
print(arr.reshape(2, 5))
print(arr.reshape(5, -1))
arr.reshape(-1, 5)
```

```python
#2.resize
arr = np.arange(10)
arr.resize(2, 5)
arr
```

```python
#3.T
arr = np.arange(12).reshape(3, 4)
print(arr)
arr.T
```

```python
#4.ravel
arr = np.arange(6).reshape(2, -1)
print(arr.ravel('F'))
arr.ravel()
```

```python
#5.flatten
a = np.floor(10 * np.random.random((3, 4)))
print(a)
a.flatten()
```

```python
#6.squeeze
arr = np.arange(3).reshape(3, 1)
print(arr.shape)
arr.squeeze().shape
```

```python
#7.transpose
arr2 = np.arange(24).reshape(2, 3, 4)
print(arr2.shape)
arr2.transpose(1, 2, 0).shape
```

```python
#1.4.2合併數組
#1.append
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
np.append(a, b)
```

```python
#2.concatenate
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6]])
print(np.concatenate((a, b), axis=0))
print(np.concatenate((a, b.T), axis=1))
```

```python
#3.stack
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
np.stack((a, b), axis=0)
```

```python
#1.7廣播機制
A = np.arange(0, 40, 10).reshape(4, 1)
B = np.arange(0, 3)
A + B
```

```python
#定义 u=f(x)=3x2−4x
%matplotlib inline
import numpy as np
from IPython import display
from d2l import torch as d2l


def f(x):
    return 3 * x ** 2 - 4 * x

#通过令x=1并让h接近0，(2.4.1)中f(x+h)−f(x)h的数值结果接近2。虽然这个实验不是一个数学证明，但我们稍后会看到，当x=1时，导数u′是2
def numerical_lim(f, x, h):
    return (f(x + h) - f(x)) / h

h = 0.1
for i in range(5):
    print(f'h={h:.5f}, numerical limit={numerical_lim(f, 1, h):.5f}')
    h *= 0.1
```
